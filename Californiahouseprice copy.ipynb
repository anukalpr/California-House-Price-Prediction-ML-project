FEATURE SCALLING

```python
# Normalisation
import numpy as np
u=np.mean(features,axis=0)
std=np.std(features,axis=0)
train_set_feature=((train_set_feature-u)/std)
test_set_feature=((test_set_feature-u)/std)
```

```python
train_set_feature.shape,test_set_feature.shape
```

```python
print(train_set_feature)
```

```python
train_set_lebal=np.array(train_set_lebal)
test_set_label=np.array(test_set_label)
```

```python
ones_train = np.ones((train_set_feature.shape[0], 1))
ones_train.shape
train_set_feature = np.hstack((ones_train, train_set_feature))
```

```python
train_set_feature.shape,test_set_feature.shape
```

SELECTION OF DESIRE MODAL FOR TRAINING DATASET

Hypothesis function

```python
def hypothesis(X,W):
    return np.dot(X,W)
```

Loss Function/Cost Function

```python
def loss_(X,Y,W):
    return np.mean((hypothesis(X, W) - Y) ** 2)
```

Gradient Function

```python
def Gradient(X,Y,W):
    m,n=X.shape
    return np.dot(X.T, (hypothesis(X, W) - Y)) / m
```

Gradient Descent Function

```python
def Gradient_descent(X,Y,learning_rate):
    eposch=200
    m,n=X.shape
    W=np.zeros((n,))
    cost=[]
    for i in range(eposch):
        Grad=Gradient(X,Y,W)
        e=loss_(X,Y,W)
        cost.append(e)
        W-=learning_rate*Grad
    return W,cost
```

```python
W,cost=Gradient_descent(train_set_feature,train_set_lebal,learning_rate=0.01)
```

Make predictions

```python
y_pred_train = hypothesis(train_set_feature, W)
```

```python
for i in range(len(y_pred_train)):
    print(y_pred_train[i],"  ,  ",train_set_lebal[i])
```

```python
ones_test = np.ones((test_set_feature.shape[0], 1))
ones_test.shape
test_set_feature = np.hstack((ones_test, test_set_feature))
```

```python
y_pred_test = hypothesis(test_set_feature, W)
```
